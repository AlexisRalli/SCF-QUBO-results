```python
import json
import numpy as np
import os
import matplotlib.pyplot as plt
import re
import pandas as pd
from symmer.operators.base import QuantumState, PauliwordOp
```

```python
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]
plt.rcParams['font.size'] = 12
```

```python
cwd = os.getcwd()
base_dir = os.path.dirname(cwd)
data_dir = os.path.join(base_dir, 'data')

classical_chem_dir = os.path.join(data_dir,
                                  'OH_classical_chem_data')
```

# Get conventional quantum chemistry package results

```python
#### GET Gaussian data

# Define regex patterns
cycle_pattern = re.compile(r'cycle\s+(\d+)')

# energy_pattern = re.compile(r'E=\s*(-?\d+\.\d+)\s+Delta-E=\s*(-?\d+\.\d+)')
energy_pattern = re.compile(r'^\s*E=\s*(-?\d+\.\d+)', re.MULTILINE)

# Read the text file
with open(os.path.join(classical_chem_dir, 'Gaussian-output1.txt'), "r") as file:
    content = file.read()

# Find all matches
cycles_guassian_1 = np.array(cycle_pattern.findall(content)).astype(int)
energies_gaussian_1 = np.array(energy_pattern.findall(content)).astype(float)
del content
```

```python
#### GET Gaussian data

# Define regex patterns
cycle_pattern = re.compile(r'Iteration\s+(\d+)')

# energy_pattern = re.compile(r'E=\s*(-?\d+\.\d+)\s+Delta-E=\s*(-?\d+\.\d+)')
energy_pattern = re.compile(r'EE=\s*(-?\d+\.\d+)', re.MULTILINE)


# Read the text file
with open(os.path.join(classical_chem_dir, 'Gaussian_output2.txt'), "r") as file:
    content = file.read()

# Find all matches
cycles_guassian_2   = np.array(cycle_pattern.findall(content)).astype(int)
energies_gaussian_2 = np.array( energy_pattern.findall(content)).astype(float)
del content
```

```python
#### GET ORCA data

# Define regex patterns

pattern = re.compile(r'^\s*(\d+)\s+(-?\d+\.\d+)', re.MULTILINE)

# Read the text file
with open(os.path.join(classical_chem_dir, 'ORCA-OH-calc.txt'), "r") as file:
    content = file.read()

matches = np.array(pattern.findall(content)[:9])

cycles_orca = np.array(matches)[:,0].astype(int)
energies_orca = np.array(matches)[:,1].astype(float)
del content
```

```python
#### GET Psi4 data

# Define regex patterns

pattern = re.compile(r'@RHF iter\s+(SAD|\d+):\s+(-?\d+\.\d+)', re.MULTILINE)

# Read the text file
with open(os.path.join(classical_chem_dir, 'psi4-output.txt'), "r") as file:
    content = file.read()

matches = np.array(pattern.findall(content))

cycles_psi4 = np.array(matches)[:,0]
cycles_psi4[0] = 0
cycles_psi4 = cycles_psi4.astype(int)
energies_psi4 = np.array(matches)[:,1].astype(float)
del content
```

```python
#### GET PySCF data

# Define regex patterns

pattern = re.compile(r'cycle=\s*(\d+)\s*E=\s*(-?\d+\.\d+)', re.MULTILINE)

# Read the text file
with open(os.path.join(classical_chem_dir, 'pyscf_output_standard.txt'), "r") as file:
    content = file.read()

matches = np.array(pattern.findall(content))

cycles_pyscf_standard = np.array(matches)[:,0].astype(int)
energies_pyscf_standard = np.array(matches)[:,1].astype(float)

del content
```

```python
#### GET PySCF data

# Define regex patterns

pattern = re.compile(r'macro=\s*(\d+)\s*E=\s*(-?\d+\.\d+)', re.MULTILINE)


# Read the text file PART 1
with open(os.path.join(classical_chem_dir, 'pyscf_soscf_part1.txt'), "r") as file:
    content = file.read()

matches = np.array(pattern.findall(content))

cycles_pyscf_soscf = np.array(matches)[:,0].astype(int)
energies_pyscf_soscf = np.array(matches)[:,1].astype(float)


# Read the text file PART 2
with open(os.path.join(classical_chem_dir, 'pyscf_soscf_part2.txt'), "r") as file:
    content = file.read()

matches = np.array(pattern.findall(content))

cycles_pyscf_soscf = np.append(cycles_pyscf_soscf, np.array(matches)[:,0].astype(int))
energies_pyscf_soscf = np.append(energies_pyscf_soscf, np.array(matches)[:,1].astype(float))

# Read the text file PART 3
with open(os.path.join(classical_chem_dir, 'pyscf_soscf_part3.txt'), "r") as file:
    content = file.read()

matches = np.array(pattern.findall(content))

cycles_pyscf_soscf = np.append(cycles_pyscf_soscf, np.array(matches)[:,0].astype(int))
energies_pyscf_soscf = np.append(energies_pyscf_soscf, np.array(matches)[:,1].astype(float))
cycles_pyscf_soscf = np.arange(1, 1+len(cycles_pyscf_soscf))
```

```python
%matplotlib inline

fig,(ax,ax2) = plt.subplots(1, 2, sharey=True, dpi=300)

# plot the same data on both axes
E_ref = energies_orca[-1]
E_ref = min(energies_gaussian_2)


ax.axhline(y=0, color='black', linestyle='--', label='Solution')
ax2.axhline(y=0, color='black', linestyle='--', label='Solution')

ax.plot(cycles_psi4+1, energies_psi4-E_ref, label='Psi4', marker='.', markersize=10, linestyle='-', linewidth=0.7, color='lightcoral') 
ax.plot(cycles_guassian_1, energies_gaussian_1-E_ref, label='Gaussian', marker='x', markersize=6, linestyle='-', linewidth=0.7, color='b') 
ax.plot(cycles_pyscf_standard, energies_pyscf_standard-E_ref, label='PySCF', marker='2', markersize=9, linestyle='-', linewidth=0.7, color='orange') 



ax2.plot(cycles_guassian_2[:20], (energies_gaussian_2-E_ref)[:20], label='Gaussian (SCF=QC)', marker='x', markersize=5,linewidth=0.7, linestyle='-', color='b') 
ax2.plot(cycles_orca+1, energies_orca-E_ref, label='ORCA', marker='H', markersize=4, linestyle='-', linewidth=0.7, color='g') 
ax2.plot(cycles_pyscf_soscf, energies_pyscf_soscf-E_ref, label='PySCF-SOSCF', marker='2', markersize=7, linestyle='-', linewidth=0.7, color='orange') 


ax.legend(loc='upper center', bbox_to_anchor=(0.7, 0.99),fancybox=False, shadow=False, ncol=1)
ax2.legend(loc='upper center', bbox_to_anchor=(0.7, 0.99),fancybox=False, shadow=False, ncol=1)

ax.set_ylabel(r'Energy Difference (Ha)')
ax.set_xlabel(r'Number of steps')
ax2.set_xlabel(r'Number of steps')

ax.set_title('Unsuccessful')
ax2.set_title('Successful')

plt.tight_layout()
fig.savefig('SCF-OH_double.tiff', dpi=300)
# fig.show()
```

# Get QUBO results

```python
qubo_data_dir = os.path.join(data_dir,
                                  'OH_data')
```

```python
alg1_file = os.path.join(qubo_data_dir, 'OH_results_6-31G.json')
with open(alg1_file, 'r') as infile:
    alg1_data = json.load(infile)
```

```python
alg1_data.keys()
```

```python
alg2_file = os.path.join(qubo_data_dir, 'ALG2_OH_results_6-31G.json')
with open(alg2_file, 'r') as infile:
    alg2_data = json.load(infile)
```

```python
alg2_data.keys()
```

```python
%matplotlib inline

fig,axes = plt.subplots(1, 2, sharey=True, dpi=300)

for idx, alg_d in enumerate([alg1_data, alg2_data]):

    # plot the same data on both axes
    E_ref = energies_orca[-1]
    E_ref = min(energies_gaussian_2)

    E_cplex = np.array(alg_d['cplex']['E_list'])
    cycles_cplex = np.arange(len(E_cplex))
    axes[idx].plot(cycles_cplex+1, E_cplex-E_ref, label='QUBO-SCF (CPLEX)', marker='.', markersize=3, 
                   linestyle='-', linewidth=0.7, color='lightcoral')
    
    E_tabu = np.array(alg_d['Tabu']['E_list'])
    cycles_tabu = np.arange(len(E_tabu))
    axes[idx].plot(cycles_tabu+1, E_tabu-E_ref, label='QUBO-SCF (Tabu)', marker='+', markersize=5, 
                   linestyle='-', linewidth=0.7, color='blue')

    E_maxcut = np.array(alg_d['maxcut']['E_list'])
    cycles_maxcut = np.arange(len(E_maxcut))
    axes[idx].plot(cycles_maxcut+1, E_maxcut-E_ref, label='MaxCut-SCF', marker='x', markersize=5, 
                   linestyle='-', linewidth=0.7, color='olive')
    
    E_anneal = np.array(alg_d['SimulatedAnnealing']['E_list'])
    cycles_anneal = np.arange(len(E_anneal))
    axes[idx].plot(cycles_anneal+1, E_anneal-E_ref, label='QUBO-SCF (annealing)', marker='1', markersize=5, 
                   linestyle='-', linewidth=0.7, color='orange')


    axes[idx].axhline(y=0, color='black', linestyle='--', label='Solution')


    axes[idx].legend(loc='upper center', bbox_to_anchor=(0.7, 0.99),fancybox=False, shadow=False, ncol=1)
    axes[idx].set_ylabel(r'Energy Difference (Ha)')
    axes[idx].set_xlabel(r'Number of steps')

    axes[idx].set_title(f'Algorithm {1+idx}')

    
    # fig.savefig('SCF-OH_double.png', dpi=300)


axes[0].set_xticks([1,2,3])
plt.tight_layout()
fig.savefig('QUBO-SCF.tiff', dpi=300)
```
