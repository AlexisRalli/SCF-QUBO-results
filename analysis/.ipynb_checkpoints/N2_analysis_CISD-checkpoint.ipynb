# Raw data for N2 study

```python
import json
import numpy as np
import os
import matplotlib.pyplot as plt
import re
import pandas as pd
```

```python
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]
plt.rcParams['font.size'] = 12
```

```python
cwd = os.getcwd()
base_dir = os.path.dirname(cwd)
data_dir = os.path.join(base_dir, 'data')

classical_chem_dir = os.path.join(data_dir,
                                  'N2_CISD')
```

```python
subdirs = [f for f in os.listdir(classical_chem_dir) if os.path.isdir(os.path.join(classical_chem_dir,
                                                            f)) if f!='log']
subdirs
```

```python
### build data dictionaries

dataframes = {}
for sub in subdirs:
    print(sub)
    sub_dir = os.path.join(classical_chem_dir,
                        sub)



    sorted_json = sorted(os.listdir(sub_dir), key=lambda s: int(re.search(r'_(\d+)\.json$', s).group(1)))
    b_lengths = np.array([int(re.search(r'_(\d+)\.json$', fname).group(1))/10 for fname in sorted_json ])

    sub_database = {'rhf_e_std_list': list(),
                    'rhf_stab_flag_list': list(),
                    'cisd_e_std_list': list(),
                    'rhf_e_newton_list': list(),
                    'rhf_newton_flag_list': list(),
                    'cisd_e_newton_list': list(),
                    
                    }
    for opt in ['maxcut', 'SimulatedAnnealing', 'Tabu', 'cplex']:
        sub_database[opt]=  {
                                'E_opt_list':       list(),
                                'rhf_boost_list':   list(),
                                'S2_list':          list(),
                                'cisd_energy_list': list()
                                }

    for idx, f_json in enumerate(sorted_json):
        json_path = os.path.join(sub_dir, f_json)
        with open(json_path, 'r') as infile:
            data = json.load(infile)

        sub_database['rhf_e_std_list']       = sub_database['rhf_e_std_list']       +  [data['scf_std_energy']]
        sub_database['rhf_stab_flag_list']   = sub_database['rhf_stab_flag_list']   +  [data['scf_std_stability']]
        sub_database['cisd_e_std_list']      = sub_database['cisd_e_std_list']      +  [data['cisd_std_energy']]
        
        sub_database['rhf_e_newton_list']      = sub_database['rhf_e_newton_list']  +  [data['newton_SCF_energy'] ]
        sub_database['rhf_newton_flag_list'] = sub_database['rhf_newton_flag_list'] +  [data['newton_SCF_stability'] ]
        sub_database['cisd_e_newton_list']      = sub_database['cisd_e_newton_list']  +  [data['cisd_newton_energy'] ]

        for opt in ['maxcut', 'SimulatedAnnealing', 'Tabu', 'cplex']:
            sub_database[opt]['E_opt_list'] =  sub_database[opt]['E_opt_list'] + [data[opt]['E_opt']]
            sub_database[opt]['rhf_boost_list'] =  sub_database[opt]['rhf_boost_list'] + [data[opt]['rhr_boost_e_tot']]
            sub_database[opt]['S2_list'] =  sub_database[opt]['S2_list'] + [data[opt]['S2_val']]

            sub_database[opt]['cisd_energy_list'] =  sub_database[opt]['cisd_energy_list'] + [data[opt]['cisd_energy']]



    scf_dataframe_data = {
    'Bond length': b_lengths,
    'SO-RHF': sub_database['rhf_e_newton_list'],
    'RHF':    sub_database['rhf_e_std_list'],
    'MaxCut': sub_database['maxcut']['rhf_boost_list'],
    'Tabu': sub_database['Tabu']['rhf_boost_list'],
    'CPLEX': sub_database['cplex']['rhf_boost_list'],
    'Annealing': sub_database['SimulatedAnnealing']['rhf_boost_list'],
    }

    cisd_dataframe_data = {
        'Bond length': b_lengths,
        'SO-RHF': sub_database['cisd_e_newton_list'],
        'RHF':    sub_database['cisd_e_std_list'],

        'MaxCut': sub_database['maxcut']['cisd_energy_list'],
        'Tabu': sub_database['Tabu']['cisd_energy_list'],
        'CPLEX': sub_database['cplex']['cisd_energy_list'],
        'Annealing': sub_database['SimulatedAnnealing']['cisd_energy_list'],
    }

    dataframes[sub] = {
                      'SCF' : scf_dataframe_data,
                      'CISD': cisd_dataframe_data
                      }
    
```

```python
%matplotlib inline
### plot SCF data

for sub in subdirs:
    key = 'SCF' 
    print(f"{sub} {key}")
    
    sub_data = dataframes[sub][key]
    ### PLOTTING 
    fig, ax_tuple = plt.subplots(2, 1, sharex=True, dpi=300)

    b_lengths = sub_data['Bond length']
    newton = np.array(sub_data['SO-RHF'])
    for idx, diff_vec in enumerate([np.zeros(len(newton)),newton ]):
    
        ax = ax_tuple[idx]
    
        if idx ==0:
            ax.plot(b_lengths, sub_data['SO-RHF'], label='SO-RHF',
                marker='2', markersize=9, linestyle='-', linewidth=0.7, color='k') 
    
        ax.plot(b_lengths, np.array(sub_data['CPLEX'])-diff_vec, label='CPLEX',
                marker='^', markersize=6, linestyle='-', linewidth=0.7, color='g') 
    
        ax.plot(b_lengths, np.array(sub_data['Tabu'])-diff_vec, label='Tabu',
                marker='H', markersize=6, linestyle='-', linewidth=0.7, color='lightblue') 
    
        ax.plot(b_lengths, np.array(sub_data['RHF']) - diff_vec, label='RHF',
                    marker='.', markersize=10, linestyle='-', linewidth=0.7, color='lightcoral') 
    
        ax.plot(b_lengths, np.array(sub_data['MaxCut'])-diff_vec, label='MaxCut',
                marker='1', markersize=6, linestyle='-', linewidth=0.7, color='orange') 
    
    
        ax.plot(b_lengths, np.array(sub_data['Annealing'])-diff_vec, label='annealing',
                marker='x', markersize=6, linestyle='-', linewidth=0.7, color='purple') 
    
    
    ax_tuple[0].set_ylabel(r'Energy (Ha)')
    ax_tuple[1].set_ylabel(r'Energy Difference (Ha)')
    ax_tuple[1].set_xlabel(r'Bond length (Å)')
    
    ax_tuple[1].set_yscale('symlog', linthresh=1e-6)
    
    ax_tuple[0].legend(loc='upper center', bbox_to_anchor=(0.73, 0.5),fancybox=False, shadow=False, ncol=2)
    
    
    plt.tight_layout()
    fig.savefig(f'{sub}.tif', dpi=300)
    fig.savefig(f'{sub}.png', dpi=300)
    
```

# Runtime info

Please note that this code has **not** been optimized to the same extent as conventional chemistry software packages. Additionally, it is implemented in Python rather than a lower-level programming language, so any runtime comparisons between methods written in different languages should be interpreted with caution.

```python
runtime_dict = dict()
for sub in subdirs:
    print(sub)

    basis_choice = sub.split('_')[1]
    sub_dir = os.path.join(classical_chem_dir,
                        sub)

    sorted_json = sorted(os.listdir(sub_dir), key=lambda s: int(re.search(r'_(\d+)\.json$', s).group(1)))
    b_lengths = np.array([int(re.search(r'_(\d+)\.json$', fname).group(1))/10 for fname in sorted_json ])

    sub_times = {'Tabu': [],
                     'cplex': [],
                     'maxcut': [],
                     'SimulatedAnnealing': []
                     }
    for idx, f_json in enumerate(sorted_json):
        json_path = os.path.join(sub_dir, f_json)
        with open(json_path, 'r') as infile:
            data = json.load(infile)

        for k in sub_times.keys():
            sub_times[k].append(data[k]['runtime'])

    sub_times['b_lengths']= b_lengths 
    runtime_dict[basis_choice] = sub_times

        
```

```python
for basis in runtime_dict.keys():

    solver_keys = list(runtime_dict[basis])
    b_key = solver_keys.pop(solver_keys.index('b_lengths'))
    b_lengths = runtime_dict[basis][b_key]
    plt.figure()
    for s in solver_keys:
        time_info = np.array(runtime_dict[basis][s]) / (60*60) # in hours
        plt.plot(b_lengths, time_info, label=s)
    plt.legend()
    plt.xlabel(r'bond length ($\AA$)')
    plt.ylabel(r'time (hours)')
    plt.title(basis)
    
```
